> ## Explain what the terms mean and discuss their interpretations.

- The bias is how systematically off target our model is. It comes from the fact that training data is just a subset of the total data. It also comes from our assumptions, and what features we have in our model. Bias is like throwing darts, where you consistently hit the same spot, but it is just not the bullseye. 
- The variance of the model comes from the fluctuations in the data, and is our error. Its again like throwing darts. Its the spread of your shots, not affected by the average value which might be the bulls eye. 
- The last term is the noise and can not possibly be explained by the model. This adds to the total error or MSE.

## Bias-Variance Analysis
We can se the effect of how complexity of the model affects the bias and variance.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
seed = 42
p_max = 50

np.random.seed(seed)
n = 100
# Make data set.
x = np.linspace(-3, 3, n).reshape(-1, 1)
y = np.exp(-x**2) + 1.5 * np.exp(-(x-2)**2) + np.random.normal(0, 0.1, x.shape)
test_error = []
train_error = []
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

for degree in range(p_max):
    model = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression(fit_intercept=False))
    clf = model.fit(X_train_scaled,y_train)
    y_fit = clf.predict(X_train_scaled)
    y_pred = clf.predict(X_test_scaled) 
    test_error.append(mean_squared_error(y_test, y_pred))
    train_error.append(mean_squared_error(y_train, y_fit))

polydegree = [_ for _ in range(p_max)]
plt.plot(polydegree, test_error, "-o", label='Test Error')
plt.plot(polydegree, train_error, "-o", label='Train Error')
plt.xlabel('Complexity [degree polynomial]')
plt.ylabel('Error [MSE]')
plt.title('Error as a Function of Complexity')
plt.legend()
plt.show()
```

> ## Discuss the bias and variance trade-off as function of your model complexity (the degree of the polynomial) and the number of data points

- As we train our model on a larger percentage of the data, the closer the model will be to the true model. This will reduce the bias, as we approach a poloynomial of degree $n$ where $n$ is the number of data points.
- The variance will decrease as we increase the number of data points, as the model will be more robust to fluctuations in the data. If too complex, the model will be overfitted, and the variance will increase. They stay separate, as the complexity grows.
